```python
import pandas as pd
```

DATA FRAMES CONCATINATION

```python
data = {
    "Name": ["Alice", "Bob", "Claire"],
    "Age": [25, 30, 27],
    "Gender": ["Female", "Male", "Female"]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

ADDING NEW ROW

```python
new_row = pd.DataFrame({"Name": "Harry","Age" : 40,"Gender": "Male"}, index=[3])
```

```python
df = pd.concat([df, new_row])
```

ADDING NEW COLUMN

```python
df["Job"] = ["Engg." , "Doctor", "Doctor", "Hacker"]
```

```python
df
```

MULTIPLE ROWS ADDITION

```python
new_rows = pd.DataFrame([{"Name": "Diljot","Age" : 40,"Gender": "Male", "Job": "Coder"},
                         {"Name": "Gaurav","Age" : 18,"Gender": "Male", "Job": "AIML Engineer"},
                         {"Name": "Harsh","Age" : 18,"Gender": "Male", "Job": "Coder"}], index=[4,5,6])
```

```python
df = pd.concat([df, new_rows])
```

```python
df
```
