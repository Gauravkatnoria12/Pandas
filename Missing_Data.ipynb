```python
import numpy as np
import pandas as pd
```

```python
data = [
    [1,2,3,4,5],
    [np.nan, 2,3,np.nan,5],
    [np.nan, 2 ,np.nan, 4, 5],
    [1, 2, 3, 4, np.nan],
]
```

```python
col = ["A", "B", "C", "D", "E"]
df = pd.DataFrame(data, columns=col)
```

```python
df
```

**Missing Values**

```python
df.isna()  #Check null values exist or not
```

```python
df.isna().sum()
```

**Deleting Missing Values + Threshold**

```python
df.dropna(thresh=4)    # The rows which have atleast 4 non null values 
```

```python
df.dropna()   # Delete the rows which have null values
```

**Fill Missing Values**

```python
df.fillna(0)  # Fill the null values
```
